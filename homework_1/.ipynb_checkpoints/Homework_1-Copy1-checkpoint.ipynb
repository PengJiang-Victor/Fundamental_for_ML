# Question 1

<h3>Why is it a good idea to standardize/normalize the predictor variables 2 and 3?</h3>

I've drawn scatter plots of our two predictors, total rooms and total bedrooms in a block, and see how they are scattered with respect to the median house value in a block. Upper two are original data of predictor 2 and 3, and lower two are predictor 2 and 3 after normalization with respect to another variable. 

Since normalization (or standardization) is essentially scaling up and down the data points of predictors without changing the relationship within the data, it's good to look at the spread of overall data points. 

Firstly, by looking at the original data points, both predictor 2 and 3 have a lot of outliers. Therefore, it's good to normalize, or scale down, these two predictors to exclude the outliers by compressing the data points. Recall that the outlier pulls the regression line towards it, so it distorts the predicted relationship that we are interested in. Secondly, total number of rooms or bed rooms are not meaningful predictors. More specifically, we assume that the median is representative of one house in the block. However, total number of rooms doesn't tell any information about individual house in one block. Thus, eventhough if we use these two predictors, it's hard to interpret the correlation between these two predictors and median house value.

<h3>why are predictor variables 4 and 5 probably not very useful by themselves to predict 
median house values in a block?</h3>

I've drawn one graph to illustrate a core problem of using just population and households.

When talking about the performance of two predictors, we need to check whether it has any collinearity, and the most straightforward way is to just plot out the scatter-plot and see if there's any linear relation between them.

Intuitively, population and households should hold a linear relationship, which means if popluation grows, the households should also grow proportionally. Observing the graph1, we can tell that there's a very clear linear relation occurs. Therefore, collinearity occurs. The coefficient of households and population will be unstable when we are doing linear regression. That is small change in the data can cause large swings in prediction of coefficients.

# Question 2

<h3>To meaningfully use predictor variables 2 (number of rooms) and 3 (number of 
bedrooms), you will need to standardize/normalize them. Using the data, is it better 
to normalize them by population (4) or number of households (5)? </h3>

I've done three linear regressions with train and test split on updated value of variables 2 and 3. To compare the effect of different ways to normalize these two variables, I print out the $R^2$, $RMSE$ for reference of performance of models with different ways of normalization.

I believe the most straightfoward way to see if change of a predictor on the performance of the model is to fit and test the model with and without that change.

It would be better to divide variables 2 and 3 by number of households. By looking at the outome, it turns out that normalizing variable 2 and 3 by population in a block has a little bit better performance as doing that by number of households in a block. Looking at the $RMSE$ of model with normalization by households, it's slightly larger (around 1100 dollars) than that of model with normalization by population, and $R^2$ of model with normalization by population is significantly larger. When combined with other predictor, this improvement could be insignificant. Moreover, in terms of interpretation and usage of the model. It would be better that if we divided total number of rooms and bed rooms by households. This all comes down to generalization of the relationship that we found within our sample. To clarify, we only collected data of specific block, but the population that we actually care about is individual house. The relationship we want to generalize is relationship between smaple statistics, which are features of an individual house. Now think about we've already had the model well trained and tested. We are predicting one specific house value by other features related to that house, and it turns out number of room per person is not a good feature of an individual house. If we predict by number of room per person, we can't obtain this data when the scale is shrunk down from block to individual house. However, room per household can be obtained by just couting the room of that house.

```python
from sklearn.model_selection import train_test_split

def rmse(y_true, y_pred):
    
    y_true, y_pred = np.array(y_true), np.array(y_pred)  # Ensure arrays
    if y_true.shape != y_pred.shape:
        raise ValueError("Input arrays must have the same shape.")
     
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

```python
data_path = "housingUnits.csv"
data = np.genfromtxt(data_path, delimiter=',', skip_header=1)
NP_X_subset = data[:, 1:3]
NP_X_subset[:, 0] = NP_X_subset[:, 0]/data[:, 3]
NP_X_subset[:, 1] = NP_X_subset[:, 1]/data[:, 3]
NP_y_subset = y
NP_X_subset[0, :]
```

```python
# Define input features (first 7 columns) and labels (y)

# Split the data into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(NP_X_subset, NP_y_subset, test_size=0.2, random_state=50)

# Fit the Linear Regression model on training data
model = LinearRegression().fit(X_train, y_train)

# Compute R^2 for training and test data
rSqr_train = model.score(X_train, y_train)
rSqr_test = model.score(X_test, y_test)

print(f"Train R^2: {rSqr_train:.3f}")
print(f"Test R^2: {rSqr_test:.3f}")

# Predict house prices for both training and test sets
yHat_train = model.predict(X_train)
yHat_test = model.predict(X_test)
RMSE_train = rmse(yHat_train, y_train)
RMSE_test = rmse(yHat_test, y_test)
print(f"Train RMSE: {RMSE_train}")
print(f"Test RMSE: {RMSE_test}")

# Scatter plot: actual vs. predicted for the test set
plt.scatter(yHat_test, y_test, alpha=0.75, marker='o')
plt.xlabel("Predicted median value")
plt.ylabel("Actual median value")
plt.title(f"Test Set: R^2 = {rSqr_test:.3f}")
plt.show()
```

# Question 3

<h3>Which of the seven variables is most *and* least predictive of housing value, from a 
simple linear regression perspective?</h3>

I did linear regression with each predictor variable on the median house value, and for that I did a train-test split. Then I print out all $RMSE$ and $R^2$ for each regression model with different predictor. I also did a linear regression on the entire set of data with the least effective predictor, and I plot actual outcomes with predicted outcomes to see what's the problem.

Since $RMSE$ and $R^2$ are two things I need to look at when accessing the performance of the model, it's the most straightforward way of doing this.

$RMSE$ tells the overall error the model introduce. Therefore, smaller the better. $R^2$ is how much the variance of outcome is explained by the model. This value should be moderate. If it's too large, like close to one, we need to check the $R^2$ of training data to see if it's overfit. If it's too small, then the model could be too simple, it could be underfit. We can see that predictor with index 5 in my output array, which is predictor 6 the median income, has highest $R^2$ and not too large and has lowest $RMSE$. Therefore, median income is the best predictor. The least effective predictor should be population which has largest $RMSE$ and extremely smallest $R^2$. For the least effective predictor, we can see that for all the predicted value, the actual outcomes are spread across all possible outcomes, and with a extremely low $R^2$, so model predicted by population alone underfits the data. Or fail to capture the pattern of data. Without constraint on our data. That is we can see the income of each household. We can actually do better job. The median income ingnore some information about individual house that we truely care about. Like how poor or rich the hosehold is, it may vary with different household. With better data, we will know additional information about dispersion of the income.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

def rmse(y_pred, y_true):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def OLS_stats(X_subset, y_subset):
    X_train, X_test, y_train, y_test = train_test_split(X_subset, y_subset, test_size=0.2, random_state=42)
    model = LinearRegression().fit(X_train, y_train)
    rSqr_test = model.score(X_test, y_test)
    yHat_test = model.predict(X_test)
    RMSE_test = rmse(yHat_test, y_test)
    return RMSE_test, rSqr_test

# Load data
data_path = "housingUnits.csv"
data = np.genfromtxt(data_path, delimiter=',', skip_header=1)

# Normalize columns
for i in [1, 2]:
    data[:, i] = data[:, i] / data[:, 4]

X = data[:, :7]  # Predictor variables (columns 0-6)
y = data[:, -1]  # Target variable (last column)

# Store RMSE and R^2 values for each predictor
rmse_values = []
r2_values = []
predictor_labels = [f"Predictor {i}" for i in range(7)]

for i in range(7):
    X_subset = X[:, i].reshape(-1, 1)  # Use each predictor separately
    RMSE_test, rSqr_test = OLS_stats(X_subset, y)
    rmse_values.append(RMSE_test)
    r2_values.append(rSqr_test)

# Print RMSE and R^2 values in readable format
print("RMSE values:")
for label, value in zip(predictor_labels, rmse_values):
    print(f"{label}: {value:.4f}")

print("\nR^2 values:")
for label, value in zip(predictor_labels, r2_values):
    print(f"{label}: {value:.4f}")

# Model: Size (in sq ft) and Age (in years)
X_subset = data[:,3].reshape(-1, 1)
model = LinearRegression().fit(X_subset,y)
rSqr = model.score(X_subset,y)

# Visualize: actual vs. predicted income (from model)
yHat = model.predict(X_subset)
plt.plot(yHat,y,'o',markersize=.75) 
plt.xlabel('Prediction from model') 
plt.ylabel('Actual price')  
plt.title('R^2 = {:.3f}'.format(rSqr))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load data
data_path = "housingUnits.csv"
data = np.genfromtxt(data_path, delimiter=',', skip_header=1)

# Normalize columns
for i in [1, 2]:
    data[:, i] = data[:, i] / data[:, 4]

X_subset = data[:, :7]
y_subset = data[:, 7]


X_train, X_test, y_train, y_test = train_test_split(X_subset, y_subset, test_size=0.2, random_state=42)
model = LinearRegression().fit(X_train, y_train)

# Predictions
yHat_train = model.predict(X_train)
yHat_test = model.predict(X_test)

# Performance metrics
rSqr_test = model.score(X_test, y_test)
RMSE_test = rmse(yHat_test, y_test)
rSqr_train = model.score(X_train, y_train)
RMSE_train = rmse(yHat_train, y_train)

# Plot predictions vs actual values
plt.figure(figsize=(10, 5))
plt.scatter(yHat_train, y_train, alpha=0.5, label='Train Data')
plt.scatter(yHat_test, y_test, alpha=0.5, label='Test Data', color='red')
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("Predicted vs Actual Values")
plt.legend()
plt.show()

print(f"RMSE_test: {RMSE_test}")
print(f"RMSE_train: {RMSE_train}")
print(f"RMSE_test: {rSqr_test}")
print(f"RMSE_test: {rSqr_train}")
```

```python
from sklearn.model_selection import train_test_split

# Load dataset
data_path = "housingUnits.csv"
data = np.genfromtxt(data_path, delimiter=',', skip_header=1)
data[:, 1] = data[:, 1]/data[:, 4]
data[:, 2] = data[:, 2]/data[:, 4]

# Define input features (first 7 columns) and labels (y)
X_subset = data[:, :1]
y_subset = y
# Swap column 0 and column 2 (for example)


def rmse(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)  # Ensure arrays
    if y_true.shape != y_pred.shape:
        raise ValueError("Input arrays must have the same shape.")
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def OLS_stats(X_subset, y_subset):
    X_train, X_test, y_train, y_test = train_test_split(X_subset, y_subset, test_size=0.2, random_state=42)
    model = LinearRegression().fit(X_train, y_train)
    rSqr_test = model.score(X_test, y_test)
    yHat_test = model.predict(X_test)
    RMSE_test = rmse(yHat_test, y_test)
    return np.array([RMSE_test, rSqr_test])

MSE_arr = np.array([])
R_2_arr = np.array([])
feature_labels = ["median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income", "ocean proximity"]

for i in range(1, 8):
    X_subset = data[:, :i]
    stats = OLS_stats(X_subset, y_subset)
    MSE_arr = np.append(MSE_arr, stats[0])
    R_2_arr = np.append(R_2_arr, stats[1])

# Plot MSE Trend
plt.figure(figsize=(8, 5))
plt.plot(range(1, 8), MSE_arr, marker='o', linestyle='-', label="MSE")
for i, txt in enumerate(feature_labels):
    plt.annotate(txt, (i+1, MSE_arr[i]), textcoords="offset points", xytext=(0, 5), ha='center')
plt.xlabel("Number of Features")
plt.ylabel("MSE")
plt.title("MSE Trend")
plt.legend()
plt.grid()
plt.show()

# Plot R^2 Trend
plt.figure(figsize=(8, 5))
plt.plot(range(1, 8), R_2_arr, marker='o', linestyle='-', label="R²")
for i, txt in enumerate(feature_labels):
    plt.annotate(txt, (i+1, R_2_arr[i]), textcoords="offset points", xytext=(0, 5), ha='center')
plt.xlabel("Number of Features")
plt.ylabel("R² Score")
plt.title("R² Trend")
plt.legend()
plt.grid()
plt.show()

data[0, :]
```
